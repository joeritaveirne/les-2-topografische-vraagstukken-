## 2.1 Forward Cut

### 2.1.2 Write The function

```python
def intersection(A : np.array,B : np.array,alpha : float, beta : float) -> np.array:
    # Converteer gon naar radialen
    alpha_rad = math.radians(alpha * 360/400)
    beta_rad = math.radians(beta * 360/400)

    # Bereken de lengte van ab
    lab = math.sqrt((B[0] - A[0])**2 + (B[1] - A[1])**2)

    # Bereken de 3e hoek van de driehoek: alpha + beta + gamma = 200gon
    gamma_gon = 200 - alpha - beta
    gamma_rad = math.radians(gamma_gon * 360/400)

    # de kaarthoek van ab
    gab = np.arctan2(B[0] - A[0], B[1] - A[1])

    # bereken de lengte van ap met de sinusregel
    lap = lab * math.sin(beta_rad) / math.sin(gamma_rad)

    # de kaarthoek van ap
    gap = gab - alpha_rad

    # Bereken co√∂rdinaten van P
    Ep = A[0] + lap * math.sin(gap)
    Np = A[1] + lap * math.cos(gap)

    return np.array([Ep, Np])
```

### 2.1.3 Executing the function

```python
A = np.array([0.0, 10.0])
B = np.array([6.0, 2.0])
alpha = 71
beta = 95

p1 = intersection(A,B,alpha, beta)
print("The first point equals:", p1)

A2 = np.array([-10.0, -3.0])
B2 = np.array([0.0, 0.0])
alpha = 132
beta = 23

p2 = intersection(A2,B2,alpha, beta)
print("The second point equals:", p2)
```
